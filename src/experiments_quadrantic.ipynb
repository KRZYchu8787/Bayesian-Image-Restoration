# How big sigma can we go?

```python
sigmas = np.arange(1, 60, 1)
mse_maps = [None for _ in sigmas]
mse_mmss = [None for _ in sigmas]
for sigma in sigmas:
    np.random.seed(seed)  # Reset seed for each iteration
    _, _, mse_map, mse_mms = process_image_quadrantic(path_file, sigma=sigma, do_not_show=True, n_iter=10)
    mse_maps[sigma-1] = mse_map
    mse_mmss[sigma-1] = mse_mms

plt.figure(figsize=(10, 5))
plt.plot(sigmas, mse_maps, label='MSE Map')
plt.plot(sigmas, mse_mmss, label='MSE MMS')
plt.xlabel('Sigma')
plt.ylabel('MSE')
plt.title('MSE vs Sigma')
plt.legend()
plt.grid()
plt.show()
```

# How many iterations do we need?

```python
n_iters = np.arange(1, 40, 1)
mse_maps_iters = [None for _ in n_iters]
mse_mmss_iters = [None for _ in n_iters]
for n_iter in n_iters:
    np.random.seed(seed)  # Reset seed for each iteration
    _, _, mse_map, mse_mms = process_image_quadrantic(path_file, sigma=10, do_not_show=True, n_iter=n_iter)
    mse_maps_iters[n_iter-1] = mse_map
    mse_mmss_iters[n_iter-1] = mse_mms
# on the same plot
plt.figure(figsize=(10, 5))
plt.plot(n_iters, mse_maps_iters, label='MSE Map')
plt.plot(n_iters, mse_mmss_iters, label='MSE MMS')
plt.xlabel('Number of Iterations')
plt.ylabel('MSE')
plt.title('MSE vs Number of Iterations')
plt.legend()
plt.grid()
plt.show()
```

# Cooling Schedule

```python
coolings = np.arange(1.01, 1.6, 0.01)
mse_maps_coolings = [None for _ in coolings]
mse_mmss_coolings = [None for _ in coolings]
for i, cooling in enumerate(coolings):
    np.random.seed(seed)  # Reset seed for each iteration
    _, _, mse_map, mse_mms = process_image_quadrantic(path_file, sigma=10, do_not_show=True, n_iter=10, cooling=cooling)
    mse_maps_coolings[i] = mse_map
    mse_mmss_coolings[i] = mse_mms
# on the same plot
plt.figure(figsize=(10, 5))
plt.plot(coolings, mse_maps_coolings, label='MSE Map')
plt.plot(coolings, mse_mmss_coolings, label='MSE MMS')
plt.xlabel('Cooling Factor')
plt.ylabel('MSE')
plt.title('MSE vs Cooling Factor')
plt.legend()
plt.grid()
plt.show()
```

# Random Search for Parameters

```python
from scipy.stats import uniform
def random_search_quadratic(n_trials=200):
    best_mse = float('inf')
    best_params = None
    for _ in range(n_trials):
        sigma = 10
        n_iter = int(uniform.rvs(5, 30))
        lam = uniform.rvs(0.05, 0.2)
        alpha = uniform.rvs(0.01, 0.2)
        beta_init = uniform.rvs(0.3, 1.6)
        cooling = uniform.rvs(1.01, 0.3)
        _, _, mse_map, mse_mms = process_image_quadrantic(
            path_file, sigma=sigma, n_iter=n_iter, lam=lam,
            alpha=alpha, beta_init=beta_init, cooling=cooling,
            do_not_show=True
        )

        if mse_map < best_mse:
            best_mse = mse_map
            best_params = (sigma, n_iter, lam, alpha, beta_init, cooling)

    return best_params, best_mse
best_params, best_mse = random_search_quadratic(n_trials=100)
print("Best Parameters:")
print(f"Sigma: {best_params[0]}, N_iter: {best_params[1]}, Lambda: {best_params[2]}, Alpha: {best_params[3]}, Beta_init: {best_params[4]}, Cooling: {best_params[5]}")
print(f"Best MSE: {best_mse}")
```
